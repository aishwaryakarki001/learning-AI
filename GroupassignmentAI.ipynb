```python
#Question 8
#Importing necessary library
import pandas as pd
import math
```

```python
#Reading CSV
df = pd.read_csv("Question8.csv")
```

```python
#Displaying dataframe
df
```

```python
#Calculating mean for each column
meanx1= df.loc[:,'x1'].mean()
meanx2= df.loc[:,'x2'].mean()
```

```python
#Applying mean value to row with null values for each column
df['x1'].fillna(meanx1, inplace=True)
df['x2'].fillna(meanx2, inplace=True)
```

```python
#Viewing dataframe 
df
```

```python
#Function to Calculate y' 
def CalculateY(row):
    return 2.7 * row['x1'] - 1.6 * row['x2'] + 0.87
```

```python
#Assigning the value to the y' using function
df["y'"]= df.apply(CalculateY,axis=1)
```

```python
df
```

```python
#Calculating residual value and assigning it to the new column
df['Residual'] = df["ycomp"] - df["y'"]
```

```python
#Calculating squared residual value and assigning it to the new column
df['SquaredResidual'] = df['Residual'] ** 2
```

```python
df
```

```python
#Calculate SSR
SSR = sum(df['SquaredResidual'])
```

```python
#Calculate MSE
MSE = SSR / 8
```

```python
#Calculate RMSE
RMSE = math.sqrt(MSE)
```

```python
#Creating new column and assigning absolute value of residual
df['Absolute Residual'] = abs(df['Residual'])
```

```python
df
```

```python
#Calculating MAE
MAE = sum(df['Absolute Residual']) / 8
```

```python
#Displaying all calculated values
print("SSR :", SSR)
print("MSE :", MSE)
print("RSME :", RMSE)
print("MAE :", MAE)
```
